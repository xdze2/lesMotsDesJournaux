```python
import pandas as pd
```

```python
print( '   - Count by day -' )

# -- load --
filename = './data_rss/count_by_day.json'
data_by_day = json.loads(open(filename).read())

print( 'nombre de mots: %i' % len(data_by_day) )

nbrs_mot_parJour = {}
for daycount in data_by_day.values():
    for day, c in daycount.items():

        if day in nbrs_mot_parJour:
            nbrs_mot_parJour[day] += c
        else:
            nbrs_mot_parJour[day] = c

print( 'nombre de jours: %i' % len(nbrs_mot_parJour) )
```

```python
from datetime import datetime
```

```python
def getDate( day ):
    date = datetime.strptime(day, '%d-%m-%Y')
    return date
```

```python
dates_v = []
nMots_v = []

for day, c in nbrs_mot_parJour.items():
    dates_v.append( getDate( day ) )
    nMots_v.append( c )
    
nMots_serie = pd.Series( nMots_v, index=dates_v )
#nMots_serie = nMots_serie[ nMots_serie.index > start_date ]
```

```python
nMots_serie.plot(figsize=(12, 4) )
plt.show()
```

```python
nMots_serie.mean()
```

```python
start_date  = datetime.strptime('16-03-2016', '%d-%m-%Y')

from matplotlib.dates import DateFormatter
formatter = DateFormatter('%d/%m')


def plotTimeLine( mot ):
    dates_v = []
    values_v = []
    #mot = 'France'
    for day, c in data_by_day[mot].items():
        dates_v.append( getDate( day ) )
        values_v.append( c )

    mot_serie = pd.Series( values_v, index=dates_v )

    s = mot_serie / nMots_serie * 1000
    s =  s.fillna(0)
    
    s = s[ s.index > start_date ]
    
    #ax = s.plot(figsize=(10, 4) )
    plt.figure(figsize=(10,2.5))
    plt.plot( s.index, s , linewidth=2 )
    plt.title(mot)
    plt.axhline(y=s.mean(), color='r', linestyle='--',  linewidth=1.2 )
    plt.axhline(y=s.mean()+s.std(), color='r', linestyle=':')
    
    if s.min() < s.mean()-s.std():
        plt.axhline(y=s.mean()-s.std(), color='r', linestyle=':')
    if s.max() > s.mean()+2*s.std():
        plt.axhline(y=s.mean()+2*s.std(), color='r', linestyle=':')
    if s.max() > s.mean()+3*s.std():
        plt.axhline(y=s.mean()+3*s.std(), color='r', linestyle=':')
        
    plt.axhline(y=s.median(), color='k', linestyle=':')    
    
    plt.gcf().axes[0].xaxis.set_major_formatter(formatter)
    plt.show()
```

```python
plotTimeLine( 'Donald Trump' )
plotTimeLine( 'Trump' )
plotTimeLine( 'Valls' )
```

```python
import numpy as np
```

```python
sorted_nuplet = sorted( data.items(), key=lambda x:len(x[0]), reverse=True )
top_mots = [x[0] for x in sorted_nuplet if x[1]>10]
```

```python
glob_stat = {}

for mot in top_mots:
    values_v = []
    for day, c in data_by_day[mot].items():
        values_v.append( c/float( nbrs_mot_parJour[day] ) )
    glob_stat[mot] = {'mean':np.mean(values_v), 'std':np.std(values_v), 'max':np.max(values_v), 'median':np.median(values_v)} 
    glob_stat[mot]['count'] = sum( data_by_day[mot].values() )
    
print len(glob_stat)
```

```python
mesures_name = glob_stat.values()[0].keys()
mesures = {}
for m in mesures_name:
    mesures[m] = [x[m] for x in glob_stat.values() ]
```

```python
def plotXY(x, y ):
    plt.figure()
    #plt.subplot(311)
    plt.loglog( mesures[x], mesures[y], 'x')
    plt.xlabel(x)
    plt.ylabel(y)
    #plt.axis('equal')
    plt.show()
```

```python
plotXY( 'mean', 'median' )
```
